```python
from interpolant import *
import numpy as np
import matplotlib.pyplot as plt

```

Let's look at the **Runge Function**
$$
f(x)=\frac{1}{1+25x^2}
$$ 
on the domain (-1,1).
It has graph:

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
plotpt=np.linspace(-1,1,1001)
plt.plot(plotpt,runge(plotpt))
```

We interpolate the Runge function at $2n+1$ equispaced points, starting from $n=1$, on the interval (-1,1)

```python
n=10
plt.plot(plotpt,runge(plotpt))
for loop in range(3,2*n+2,2):
    datax=np.linspace(-1,1,loop)
    datay=runge(datax)
    poly=Interpolant(datax,datay)
    plt.scatter(datax,datay,c='r')
    plt.plot(plotpt,poly.eval(plotpt))
```

```python
for n in range(10,30,5):
    datax=np.cos((2*np.arange(n)+1)*np.pi/(2.0*n))
    plt.plot(plotpt,runge(plotpt))
    datay=runge(datax)
    poly=Interpolant(datax,datay)
    plt.scatter(datax,datay,c='r')
    plt.plot(plotpt,poly.eval(plotpt))
```

```python
from scipy.special import cbrt
for n in range(10,31,10):
    datax=np.cos((2*np.arange(n)+1)*np.pi/(2.0*n))
    datay=np.absolute(datax)
    #datay=cbrt(datax)
    poly=Interpolant(datax,datay)
    plt.scatter(datax,datay,c='r')
    plt.plot(plotpt,poly.eval(plotpt))
    plt.plot(plotpt,np.absolute(plotpt))
    #plt.plot(plotpt,cubrt(plotpt))
```

```python
from scipy.interpolate import CubicSpline

n=60
plt.rcParams['figure.figsize'] = 10, 10
datax=np.cos((2*np.arange(n)+1)*np.pi/(2.0*n))
datax=np.sort(datax)
datay=cbrt(datax)
plt.scatter(datax,datay,c='r')
poly=Interpolant(datax,datay)
plt.plot(plotpt,poly.eval(plotpt))
plt.plot(plotpt,cbrt(plotpt))
datax2=np.linspace(-1,1,n)
datay2=cbrt(datax2)
cs=CubicSpline(datax2,datay2)
plt.plot(plotpt,cs(plotpt))
plt.scatter(datax2,datay2,c='b')
```
